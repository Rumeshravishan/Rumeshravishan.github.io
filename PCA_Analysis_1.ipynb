```python
import numpy as np
import matplotlib.pyplot as plt

data = np.array([[40,20],
                [55,30],
                [70,60],
                [50,35],
                [45,40],
                [62,75],
                [45,30],
                [68,80],
                [80,70],
                [75,90]])  # Left side is maths and right side is science marks

plt.scatter(data[:,0],data[:,1]) # create a scatter plot, arrays slicing technique have been used here
plt.xlabel("Maths Marks")
plt.ylabel("Science Marks")
plt.show()

# Do feature scalling x begins from 40 and then y begins from 20

from sklearn.preprocessing import StandardScaler # library for feature scalling
scaler = StandardScaler()
scaled_data =scaler.fit_transform(data)
print(scaled_data)

plt.scatter(scaled_data[:,0],scaled_data[:,1]) # create a scatter plot, arrays slicing technique have been used here
plt.xlabel("Maths Marks")
plt.ylabel("Science Marks")
plt.show()


```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(scaled_data)
```

```python
pca.explained_variance_ #data spread from each component (straight line)
```

```python
pca.explained_variance_ratio_ #take the ratio of above fact
```

```python
pca = PCA(n_components=1)  # reduce dimention from two dimentions to one
pca.fit(scaled_data)
pca.explained_variance_
```

```python

pca_scaled_data = pca.transform(scaled_data)  # after reducing data
pca_scaled_data.shape
```

```python
scaled_data.shape
```

```python
pca_scaled_data = pca.inverse_transform(pca_scaled_data)
plt.scatter(scaled_data[:,0],scaled_data[:,1])
plt.scatter(pca_scaled_data[:,0],pca_scaled_data[:,1])
```
